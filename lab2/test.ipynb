```python
#Import modules
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from scipy import stats
import seaborn as sns

titanic_df = pd.read_csv('./data/Titanic.csv', index_col='PassengerId',sep=',')

#Lets check for null and nan values in this data set
titanic_df.isna().sum()
# You can also use df.isnull().sum()
```

```python
#Filling Age, Fare with median value
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df['Fare'].fillna(titanic_df['Fare'].median(), inplace=True)

#Filling Embarked with 'S'
titanic_df['Embarked'].fillna('S', inplace=True)
```

```python
# Check the DataFrame after filling missing values
print("\nDataFrame After Filling Missing Values:")
print(titanic_df)

# Check for missing values again
print("\nMissing Values After Filling:")
print(titanic_df.isna().sum())
```

```python
#We can check the numerical difference of mean and median
# mean age
print('Mean of "Age" is %.2f' %(titanic_df["Age"].mean(skipna=True)))
# median age
print('Median of "Age" is %.2f' %(titanic_df["Age"].median(skipna=True)))
```

```python
# percent of missing "Cabin" 
print('Percent missing in "Cabin" column: %.2f%%' %((titanic_df['Cabin'].isnull().sum()/titanic_df.shape[0])*100)) 
```

```python
#Add a new column for HasCabin
# ~ invert the bool value, so that no cabin = False
titanic_df['HasCabin'] = ~titanic_df.Cabin.isnull()
titanic_df.head()
```

```python
#Lets add a new column called Title (regex101.com)
titanic_df['Title'] = titanic_df.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1)) 
```

```python
sns.countplot(x='Title', data=titanic_df, hue='Title')
plt.xticks(rotation=45)
```

```python
titanic_df["Title"].value_counts()
```

```python
# Creating numerical columns for age and fare
# qcut is used to discretize variables into equal-sized buckets, in order to create a categorical variable
# Think of it as creating 4 bins of age range 0, 1, 2, 3
titanic_df['CatAge'] = pd.qcut(titanic_df.Age, q=4, labels=False)
titanic_df['CatFare']= pd.qcut(titanic_df.Fare, q=4, labels=False)
titanic_df.head()
```

```python
#Dropping columns that wont make sense in binary format
titanic_df = titanic_df.drop(["Age", "Fare", "Cabin", "Name", "Ticket"], axis=1)
titanic_df.head()
```

```python
# Converting to binary values
titanic_df_dum = pd.get_dummies(titanic_df, drop_first=True)
titanic_df_dum.head()
```

```python
titanic_df_dum.to_csv('data/Titanic_Cleaned.csv', index=False)
```
